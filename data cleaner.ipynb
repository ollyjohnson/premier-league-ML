```python
import pandas as pd
```

```python
matches = pd.read_csv('premier-league-matches.csv')
matches2024 = pd.read_csv('matches.csv')
```

```python
matches.head()
```

```python
matches2024.head()
```

```python
matches2024 = matches2024[matches2024['Venue'] != 'Away']
matches2024 = matches2024.sort_values(by='Round')
new_rows = []
for match in matches2024.iterrows():
    result = "D"
    if match[1]['Result'] == 'W':
        result = "H"
    elif match[1]['Result'] == 'L':
        result = "A"
    round = match[1]['Round'].replace("Matchweek ", "")
    date = match[1]['Date'].replace("-", "/")
    new_rows.append(['2024', round, date, match[1]['Team'], match[1]['GF'], match[1]['GA'], match[1]['Opponent'], result])

new_rows_df = pd.DataFrame(new_rows, columns=['Season_End_Year', 'Wk', 'Date', 'Home', 'HomeGoals', 'AwayGoals', 'Away', 'FTR'])
matches = pd.concat([matches, new_rows_df], ignore_index=True)
```

```python
matches['Wk'] = matches['Wk'].astype(int)
matches = matches.sort_values(by=['Season_End_Year', 'Wk'])
matches.to_csv('clean-matches.csv', index=False)
```
